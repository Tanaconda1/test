### pywaveletsライブラリ

```python
import pywt
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
coef, freqs=pywt.cwt(y,np.arange(1,129),'gaus1')
```

```python
np.arange(1,129)
```

```python
coefT=np.transpose(coef)
```

```python
plt.matshow(coefT) # doctest: +SKIP
plt.show() # doctest: +SKIP
```

### ２つめのプログラム

```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
```

```python
sig  = y.copy()
```

```python
len(sig)
```

```python
widths = np.arange(1, 31)
cwtmatr, freqs = pywt.cwt(sig, widths, 'mexh')
```

```python
cwtmatrT = np.transpose(cwtmatr)
```

```python
plt.imshow(cwtmatrT, extent=[-1, 1, 1, 31], cmap='PRGn', aspect='auto', vmax=abs(cwtmatrT).max(), vmin=-abs(cwtmatrT).max())  # doctest: +SKIP
plt.show() # doctest: +SKIP
```
